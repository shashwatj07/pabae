```python
# greedy = pd.read_csv("/scratch/sharma96/ab/abae/experiments/results/mse/CelebARecords_mse_100_egreedy.csv", index_col=0)
orig = pd.read_csv("/scratch/sharma96/ab/abae/experiments/results/mse/CelebARecords_mse_ada_ci.csv", index_col=0)
imp = pd.read_csv("/scratch/sharma96/ab/abae/experiments/results/mse/CelebARecords_mse_imp.csv", index_col=0)
ucb = pd.read_csv("/scratch/sharma96/ab/abae/experiments/results/mse/CelebARecords_mse_ucb.csv", index_col=0)
ada = pd.read_csv("/scratch/sharma96/ab/abae/experiments/results/mse/CelebARecords_mse_ada.csv", index_col=0)

imp.drop(imp.columns[0], axis=1, inplace=True)
orig.drop(orig.columns[0], axis=1, inplace=True)
ada.drop(ada.columns[0], axis=1, inplace=True)
ucb.drop(ucb.columns[0], axis=1, inplace=True)
```

```python
def rmse(x, y):
    return pd.Series(np.sqrt(np.mean((x-y)**2)))

def return_errors(df):
    rmsedf = pd.DataFrame()

    for algo in df.columns:
        rmsedf[algo] = rmse(df["truth"], df[algo])

    our_y = []
    our_x = []
    uni_y = []
    uni_x = []
    for algo in rmsedf:
        if "uniform" in algo:
            uni_y.append(rmsedf[algo].mean())
            uni_x.append(algo.split("_")[1])
        if "ours" in algo:
            our_y.append(rmsedf[algo].mean())
            our_x.append(algo.split("_")[1])
#         plot_y.append(rmsedf[algo].mean())
#         plot_x.append(algo)
    return our_y, our_x, uni_y, uni_x

def return_errors_ci(df):
    rmsedf = pd.DataFrame()

    for algo in df.columns:
        rmsedf[algo] = rmse(df["truth"], df[algo])

    our_y = []
    our_x = []
    uni_y = []
    uni_x = []
    lb_x=[]
    lb_y=[]
    ub_x=[]
    ub_y=[]
    for algo in rmsedf:
        if "uniform" in algo:
            uni_y.append(rmsedf[algo].mean())
            uni_x.append(algo.split("_")[1])
        elif "lb" in algo:
            lb_y.append(df[algo].mean())
            lb_x.append(algo.split("_")[1])
        elif "ub" in algo:
            ub_y.append(df[algo].mean())
            ub_x.append(algo.split("_")[1])
        elif "ours" in algo:
            our_y.append(rmsedf[algo].mean())
            our_x.append(algo.split("_")[1])
#         plot_y.append(rmsedf[algo].mean())
#         plot_x.append(algo)
    return our_y, our_x, uni_y, uni_x, lb_y, lb_x, ub_y, ub_x

```

```python
imp_y, imp_x, uni_y, uni_x = return_errors(imp)
orig_y, orig_x, _, _ = return_errors(orig)
ucb_y, ucb_x,_, _ = return_errors(ucb)
ada_y, ada_x,_, _ = return_errors(ada)
plt.figure(figsize=(14,4))
plt.plot(imp_x,imp_y, label="Importance Sampling")
plt.plot(ucb_x,ucb_y, label="UCB")
plt.plot(orig_x,orig_y, label="ABae")
plt.plot(ada_x,ada_y, label="Adaptive Sampling")
plt.ylabel("RMSE", fontsize=14)
plt.xlabel("Oracle Budget (n)", fontsize=14)
plt.title("Performance on CelebA", fontsize=16)
plt.legend(fontsize=12)
plt.grid()
plt.show()
```

```python
#adaptive scores with ci
celeba = pd.read_csv("/scratch/sharma96/ab/abae/experiments/results/mse/CelebARecords_mse_ada_ci.csv", index_col=0)
jackson = pd.read_csv("/scratch/sharma96/ab/abae/experiments/results/mse/jackson_mse_ada_ci.csv", index_col=0)
taipei = pd.read_csv("/scratch/sharma96/ab/abae/experiments/results/mse/taipei_mse_ada_ci.csv", index_col=0)
moviefacesv2 = pd.read_csv("/scratch/sharma96/ab/abae/experiments/results/mse/moviefacesv2_mse_ada_ci.csv", index_col=0)

ours_y, ours_x, uni_y, uni_x, lb_y, lb_x, ub_y, ub_x = return_errors_ci(celeba)
plt.figure(figsize=(14,4))
plt.plot(ours_x,ours_y, label="Ours")
plt.plot(ub_x,ub_y, label="Upper Bound")
plt.plot(lb_x,lb_y, label="Lower Bound")
plt.plot(uni_x,uni_y, label="Uniform")
plt.ylabel("RMSE", fontsize=14)
plt.xlabel("Oracle Budget (n)", fontsize=14)
plt.title("Performance on CelebA", fontsize=16)
plt.legend(fontsize=12)
plt.grid()
plt.show()
```
